```python
import os
import time
from bs4 import BeautifulSoup
from playwright.async_api import async_playwright, TimeoutError as PlaywrightTimeout
import nest_asyncio
import asyncio

nest_asyncio.apply()  # Jupyter Notebook에서 비동기 코드를 실행하기 위해 필요

async def fetch_html(url, selector, sleep=5, retries=3):
    html_content = None
    for attempt in range(1, retries + 1):
        time.sleep(sleep * attempt)
        try:
            async with async_playwright() as playwright:
                browser = await playwright.chromium.launch()
                page = await browser.new_page()
                await page.goto(url)
                print(await page.title())
                html_content = await page.inner_html(selector)
                await browser.close()
        except PlaywrightTimeout:
            print(f"Timeout error on {url}, attempt {attempt}")
            continue
        else:
            break
    return html_content
```

```python
async def scrape_season(season):
    base_url = f"https://www.basketball-reference.com/leagues/NBA_{season}_games.html"
    html = await fetch_html(base_url, "#content .filter")

    soup = BeautifulSoup(html, 'html.parser')
    links = soup.find_all("a")
    standings_pages = [f"https://www.basketball-reference.com{link['href']}" for link in links]

    for page_url in standings_pages:
        file_name = page_url.split("/")[-1] + ".html"
        save_path = os.path.join(STANDINGS_DIR, file_name)

        if not os.path.exists(STANDINGS_DIR):
            os.makedirs(STANDINGS_DIR)

        if os.path.exists(save_path):
            continue

        page_html = await fetch_html(page_url, "#all_schedule")
        with open(save_path, "w+", encoding='utf-8') as file:
            file.write(page_html)
```

```python
for season in SEASONS:
    await scrape_season(season)
```

```python
async def scrape_game(standings_file):
    with open(standings_file, 'r') as f:
        html = f.read()

    soup = BeautifulSoup(html)
    links = soup.find_all("a")
    hrefs = [l.get('href') for l in links]
    box_scores = [f"https://www.basketball-reference.com{l}" for l in hrefs if l and "boxscore" in l and '.html' in l]

    for url in box_scores:
        save_path = os.path.join(SCORES_DIR, url.split("/")[-1])
        if os.path.exists(save_path):
            continue

        try:
            html = await fetch_html(url, "#content")
        except PlaywrightTimeout:
            print(f"Timeout error occurred for {url}. Skipping.")
            continue

        if not html:
            continue

        with open(save_path, "w+") as f:
            f.write(html)
```

```python
import os
import pandas as pd

SCORE_DIR = "data/scores"

# Ensure the directory exists or create it if it doesn't
if not os.path.exists(SCORE_DIR):
    os.makedirs(SCORE_DIR)

# Now you can use box_scores
box_scores = os.listdir(SCORE_DIR)
box_scores = [os.path.join(SCORE_DIR, f) for f in box_scores if f.endswith(".html")]

for season in SEASONS:
    files = [s for s in standings_files if str(season) in s]

    for f in files:
        filepath = os.path.join(STANDINGS_DIR, f)

        try:
            await scrape_game(filepath)
        except FileNotFoundError:
            print(f"파일이 존재하지 않습니다: {filepath}. 건너뜁니다.")
            continue
```

```python
import os
import pandas as pd

SCORE_DIR = "data/scores"
```

```python
box_scores = os.listdir(SCORE_DIR)
box_scores = [os.path.join(SCORE_DIR, f) for f in box_scores if f.endswith(".html")]
```

```python
box_scores
```

```python
from bs4 import BeautifulSoup

def parse_html(box_score):
    with open(box_score) as f:
        html = f.read()

    soup = BeautifulSoup(html)
    [s.decompose() for s in soup.select("tr.over_header")]
    [s.decompose() for s in soup.select("tr.thead")]
    return soup
```

```python
def read_season_info(soup):
    nav = soup.select("#bottom_nav_container")[0]
    hrefs = [a["href"] for a in nav.find_all('a')]
    season = os.path.basename(hrefs[1]).split("_")[0]
    return season
```

```python
def read_line_score(soup):
    line_score = pd.read_html(str(soup), attrs = {'id': 'line_score'})[0]
    cols = list(line_score.columns)
    cols[0] = "team"
    cols[-1] = "total"
    line_score.columns = cols

    line_score = line_score[["team", "total"]]

    return line_score
```

```python
def read_stats(soup, team, stat):
    df = pd.read_html(str(soup), attrs = {'id': f'box-{team}-game-{stat}'}, index_col=0)[0]
    df = df.apply(pd.to_numeric, errors="coerce")
    return df
```

```python
games = []
base_cols = None
for box_score in box_scores:
    soup = parse_html(box_score)

    line_score = read_line_score(soup)
    teams = list(line_score["team"])

    summaries = []
    for team in teams:
        basic = read_stats(soup, team, "basic")
        advanced = read_stats(soup, team, "advanced")

        totals = pd.concat([basic.iloc[-1,:], advanced.iloc[-1,:]])
        totals.index = totals.index.str.lower()

        maxes = pd.concat([basic.iloc[:-1].max(), advanced.iloc[:-1].max()])
        maxes.index = maxes.index.str.lower() + "_max"

        summary = pd.concat([totals, maxes])

        if base_cols is None:
            base_cols = list(summary.index.drop_duplicates(keep="first"))
            base_cols = [b for b in base_cols if "bpm" not in b]

        summary = summary[base_cols]

        summaries.append(summary)
    summary = pd.concat(summaries, axis=1).T

    game = pd.concat([summary, line_score], axis=1)

    game["home"] = [0,1]

    game_opp = game.iloc[::-1].reset_index()
    game_opp.columns += "_opp"

    full_game = pd.concat([game, game_opp], axis=1)
    full_game["season"] = read_season_info(soup)

    full_game["date"] = os.path.basename(box_score)[:8]
    full_game["date"] = pd.to_datetime(full_game["date"], format="%Y%m%d")

    full_game["won"] = full_game["total"] > full_game["total_opp"]
    games.append(full_game)

    if len(games) % 100 == 0:
        print(f"{len(games)} / {len(box_scores)}")
```

```python
games_df = pd.concat(games, ignore_index=True)
```
